```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Functions to read and plot the hypervolume values

```python
def read_data(folder):
    executions_number = 5
    df_list = []
    for i in range(executions_number):
        df = pd.read_csv(folder + 'hypervolume_execution_'+str(i)+'.csv',header=None)
        df.columns = str(i).split()
        df_list.append(df)

    return pd.concat(df_list, axis=1)

def plot_convergence(result):
    plt.figure(figsize=(10,8))
    result.mean(axis=1).plot()
    plt.show()
```

### Convergence for DTLZ5 with 2 objectives

```python
folder = 'DTLZ5-2/'
result = read_data(folder)
plot_convergence(result)
```

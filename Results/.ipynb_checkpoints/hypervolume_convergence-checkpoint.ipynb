```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
executions_number = 5
df_list = []
for i in range(executions_number):
    df = pd.read_csv('hypervolume_execution_'+str(i)+'.csv',header=None)
    df.columns = str(i).split()
    df_list.append(df)


result = pd.concat(df_list, axis=1)
```

```python
result.mean(axis=1).plot()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def read_data(problem):
    x = []
    y = []
    with open(problem +'/'+problem + '-solutions.csv') as file:
        for line in file:
            try:
                splitted_line = line.split('Objectives: ')[1].replace('AlgorithmAttributes: {}','')\
                    .strip().split()
                x.append(float(splitted_line[0]))
                y.append(float(splitted_line[1]))
            except:
                pass
            
    df = pd.DataFrame(data=[x,y]).T
    df.columns = 'X Y'.split()
    return df

def read_pop_ini(problem):
    x = []
    y = []
    with open(problem +'/pop_ini.csv') as file:
        for line in file:
            try:
                splitted_line = line.split('Objectives: ')[1].replace('AlgorithmAttributes: {}','')\
                    .strip().split()
                x.append(float(splitted_line[0]))
                y.append(float(splitted_line[1]))
            except:
                pass
            
    df = pd.DataFrame(data=[x,y]).T
    df.columns = 'X Y'.split()
    return df

def plot_solutions(df):
    plt.figure(figsize=(10,8))
    df.plot(kind='scatter', x= 'X', y= 'Y')
    plt.show()
```

```python
problem = 'DTLZ2-2'
plot_solutions(read_data(problem))
```

```python
problem = 'DTLZ2-2'
plot_solutions(read_pop_ini(problem))
```
